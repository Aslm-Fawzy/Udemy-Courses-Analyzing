#**Introduction**
>Udemy, Inc. is an American massive open online course (MOOC) provider aimed at professional adults and students. It was founded in May 2010 by Eren Bali, Gagan Biyani, and Oktay Caglar.
As of June of 2021, the platform has more than 44 million students, 183,000 courses, and 65,000 instructors teaching courses in 75 languages. There have been over 594 million course enrollments.

>This dataset contains 3.682 records of courses from 4 subjects (Business Finance, Graphic Design, Musical Instruments and Web Design) taken from Udemy.
Udemy is a massive online open course (MOOC) platform that offers both free and paid courses. Anybody can create a course, a business model by which allowed Udemy to have hundreds of thousands of courses.
This version modifies column names, removes empty columns and aggregates everything into a single csv file for ease of use.

> For More about Dataset : https://www.kaggle.com/datasets/andrewmvd/udemy-courses

## Data Columns ⚡
•	course_id - id field for courses

•	course_title - title field for courses

•	url - url field to course page

•	is_paid - boolean field - True for Paid / False for Free

•	price - price field for course fee

•	num_subscribers - demand field for each courses

•	num_reviews - review number for each courses

•	num_lectures - lecture per course

•	level - course level by trainee experience

•	content_duration - course duration in hours

•	published_timestamp - timestamp field for publication

•	subject - course type field

## Data Cleaning ♾
1. Checking Duplicated Rows or Columns
3. Converting  datatype of published_timestamp to datatype of datatime
4. Creating Time Features as hours , days , months and hours from published_timestamp
5. Removing unneeded Columns


###  Checking Duplicated Rows or Columns

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

### Converting datatype of published_timestamp Column to datatype of datatime

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df.info()
```

### Creating Time Features from published_timestamp Column 

```python
df['Published_Month_Name'] = df['published_timestamp'].dt.month_name()
df['Published_Day_Name'] = df['published_timestamp'].dt.day_name()
df['Published_Year'] = df['published_timestamp'].dt.year
```

```python
df['Published_date'] = pd.to_datetime(df['published_timestamp'].dt.date)
```

```python
df
```

### Removing unneeded Columns

```python
df.drop(['course_id','published_timestamp'],axis = 1 , inplace = True)
```

```python
df
```

#Exploratory Data Analysis ⛲



## Courses Price Distribustion 💰

```python
sns.set(rc={'figure.figsize':(15, 8)}, font_scale=1.3)
df['price'].plot(kind = 'hist')
plt.title('Udemy Courses Price Distribustion',size = 20,fontdict={'fontweight' : 'bold'})
```

## Free and Paid Courses Count 🆓

```python
df['is_paid'].value_counts()
```

```python
plt.pie(df['is_paid'].value_counts(),labels=['Paid','Free'],
        shadow= True, autopct='%.2f%%', explode = [0,0.2])
plt.title('Free and Paid Courses Count',size = 20,fontdict={'fontweight' : 'bold'})
plt.axis ('equal')
plt.show()
```

## Count of Courses Subjects  🔷

```python
df['subject'].value_counts()
```

```python
plt.pie(df['subject'].value_counts(),labels=df['subject'].value_counts().index,
        shadow= True, autopct='%.2f%%', explode = [0.1,0,0,0],startangle=90)
plt.title('Count of Udemy Courses Subjects',size = 20,fontdict={'fontweight' : 'bold'})
plt.axis ('equal')
plt.show()
```

## Course Subject and Pricing 

```python
labels = list(df['subject'].value_counts().index)
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(labels, df['subject'][df['is_paid']==True].value_counts(), label='Paid'
                       ,facecolor ='b',edgecolor='k')
ax.bar(labels, df['subject'][df['is_paid']==False].value_counts(), bottom=df['subject'][df['is_paid']==True].value_counts(),
       label="Free",facecolor ='r',edgecolor='k')
plt.title(' Paid and Free Courses in Each Subject',fontdict={'fontweight' : 'bold'},size = 20)
ax.legend()
```

```python
sns.barplot(x=df['subject'], y = df['price'],
              order = ['Web Development','Business Finance','Graphic Design','Musical Instruments']
            , palette='Paired_r')
plt.title('Average Price for each Course Subject',fontdict={'fontweight' : 'bold'},size = 20)
plt.show()
```

## Levels in Courses Subjects 🍃

```python
sns.countplot(x=df['level'], order = df['level'].value_counts().index)
plt.title('Count of Levels in Udemy Courses',size = 20,fontdict={'fontweight' : 'bold'})
plt.show()
```

```python
most_level_in_each_subject = df.groupby('subject').agg(lambda x:x.value_counts().index[0])['level']
most_level_in_each_subject= pd.DataFrame(most_level_in_each_subject)
most_level_in_each_subject
```
